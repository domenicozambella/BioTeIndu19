## Filtrare righe di un dataframe

Osserviamo prima che alcune espressioni danno come risultato i valori booleani `True` e `False`

```python
28 > 28
```

```python
28 >= 28
```

```python
28 == 28
```

```python
28<= 29
```

Queste operazioni posso essere vettorializzate. Possiamo confrontare un vettore con un numero, o due vettori numerici, e ottenere un vettore booleano. Per esempio. Il seguente comando da la lista delle rige in cui la temperatura massima Ã¨ maggiore di 8.

```python
df['maxtp'] > 8
```

I vettori booleani possono essere salvati in una variabile. Per esempio:

```python
max28 = df['maxtp'] > 28
```

Un vettore boolean puÃ² essere usato per filtrare determinate righe di un dataframe. Nel seguente esempio filtriamo i gioni in cui la temperatura massima Ã¨ (strttamemte) superiore a 28Â°C.

```python
df[max28]
```

Tra vettori a valori Booliani (vero o falso) possiamo fare le seguenti operazioni

`&` congiunzione (un **e**)

`|` disgiunzione (un **o** non esclusivo)

`~` negazione

Per esempio nella seguente cella isoliamo le date in cui le massime sono state superiori a 28Â°C o le minime sono state inferiori a -8Â°C.

```python
min08 = df['mintp'] < -8
df[max28 | min08]
```

Esistono vari modi di ottenere questo risultato. Un alternativa Ã¨ usare la funzione `query()`

```python
df.query( 'mintp <= -8 or maxtp >= 28' )
```

## I dati temporali

La prima colonna contiene delle date. PerÃ² per `Python` queste sono solo delle stringhe di caratteri. 

Possiamo forzare `Python` a leggerle come date.

In `Python` la libreria standard che tratta dati temporali Ã¨ `datetime`. Parti di questa libreria sono incorporati in `pandas`.

```python
df1 = df.copy()
df1['date'] = pd.to_datetime(df1['date'])
df1.tail()
```

```python
df1['date'] > '2000-01-01'
```

```python
df2 = df1.set_index('date')
df2.tail()
```
