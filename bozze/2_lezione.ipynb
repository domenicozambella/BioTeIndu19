```python
import pandas as pd
df = pd.read_csv("../dati/Cork_Airport.csv", 
                 skiprows=24, 
                 usecols=['date', 'maxtp', 'mintp'],
                )
df.tail(2)
```

## Filtrare righe di un dataframe

```python
max28 = df['maxtp'] >= 28
df[max28]
```

Tra vettori a valori Booliani (vero o falso) possiamo fare le seguenti operazioni

`&` congiunzione (un **e**)

`|` disgiunzione (un o non esclusivo)

`~` negazione

```python
min08 = df['mintp'] <= -8
df[max28 | min08]
```

```python
feb29 = df['date'].map(lambda x: '29-feb' in x)
df[feb29]
```

## I dati temporali

La prima colonna contiene delle date. Però per `Python` queste sono solo delle stringhe di caratteri. 

Possiamo forzare `Python` a leggerle come date.

In `Python` la libreria standard che tratta dati temporali è `datetime`. Parti di questa libreria sono incorporati in `pandas`.

```python
df1 = df.copy()
df1['date'] = pd.to_datetime(df1['date'])
df1
```
